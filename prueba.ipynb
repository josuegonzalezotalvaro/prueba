# SQL

```python
import pandas as pd
import sqlite3
```

## 1

```python
# 1. Leer el archivo Excel
sales_df = pd.read_excel("Base de datos BI.xlsx", sheet_name="Sales")
product_df = pd.read_excel("Base de datos BI.xlsx", sheet_name="Product")
customer_df = pd.read_excel("Base de datos BI.xlsx", sheet_name="Customers")
geography_df = pd.read_excel("Base de datos BI.xlsx", sheet_name="Geography")
```

```python
# 2. Crear conexión a base de datos SQLite
conn = sqlite3.connect("mi_base.db")
```

```python
# 3. Guardar DataFrame como tabla SQL
sales_df.to_sql("sales", conn, if_exists="replace", index=False)
product_df.to_sql("product", conn, if_exists="replace", index=False)   
customer_df.to_sql("customer", conn, if_exists="replace", index=False)
geography_df.to_sql("geography", conn, if_exists="replace", index=False) 
```

```python
# 4. Hacer una consulta SQL
query = """
SELECT geography.City, product.ProductName, COUNT(sales.ProductKey) AS ventas
FROM product
    INNER JOIN sales ON product.ProductK = sales.ProductKey
    INNER JOIN customer ON sales.CustomerKey = customer.Customer_Key
    INNER JOIN geography ON customer.Geography = geography.GeographyKey
GROUP BY geography.City
ORDER BY ventas DESC
"""

resultado = pd.read_sql_query(query, conn)

print("\nResultado de la consulta:")
print(resultado)
```

```python
# 5. Cerrar conexión
conn.close()
```
